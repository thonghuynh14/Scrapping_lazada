```python
import re

def extract_info(data):
    # Tạo các biểu thức chính quy để tìm thông tin cần thiết
    brand_pattern = re.compile(r"Thương hiệu\s+(\w+)")
    sku_pattern = re.compile(r"SKU\s+([\w-]+)")
    material_pattern = re.compile((r"Chất Liệu\s+(\w+)"))
    
    # Tìm kiếm thông tin trong chuỗi
    brand = brand_pattern.search(data)
    sku = sku_pattern.search(data)
    material = material_pattern.search(data)
    brand = brand.group(1) if brand else None,
    sku = sku.group(1) if sku else None,
    material = material.group(1).strip() if material else None
    return brand, sku, material
```

```python
df_final[['Brand', 'Mã SP', 'Chất liệu']] = df_final['feature'].apply(lambda x: pd.Series(extract_info(x)))
df_final = df_final.drop(columns=['feature'])
```

```python
df_final['price'] = df_final['price'].str.replace("₫", "").str.replace(".", "")
df_final['price'] = df_final['price'].astype(int)
```

```python
def process_quantity_improved(quantity):
    # Loại bỏ " Đã bán", dấu phẩy và dấu "+"
    quantity = quantity.replace(" Đã bán", "").replace(",", "").replace("+", "")
    # Xử lý cho trường hợp có "k"
    if "k" in quantity:
        quantity = float(quantity.replace("k", "")) * 1000
    else:
        quantity = float(quantity)
    return int(quantity)

# Áp dụng hàm đã cải thiện lên cột 'quantity_sold'
df_final['quantity_sold'] = df_final['quantity_sold'].apply(process_quantity_improved)

df_final['quantity_sold']
```

```python
# Sử dụng biểu thức chính quy để trích xuất số trước dấu "%"
df_final['sale_off'] = df_final['sale_off'].str.extract(r'(\d+)%').fillna(0).astype(int)

df_final['sale_off']
```
