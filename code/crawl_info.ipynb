```python
import requests
import re
from requests import get
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep

from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.implicitly_wait(0)
driver.maximize_window()
print("Driver setup and browser opened")
```

```python
df = pd.read_csv('wallet_lazada.csv')
df
```

```python
import random
from selenium.common.exceptions import NoSuchElementException
list_item = []
for ele in df['links'][2600:]:
    html2 = driver.get(str(ele))
    dict_item = {}
    while True:
        driver.execute_script("window.scrollBy(0, 1000);")
        sleep(2)
        try:
            button = driver.find_element(By.CSS_SELECTOR, "#module_product_detail > div > div > div.expand-button.expand-cursor > button")
            # Nhấp vào nút "XEM THÔNG SỐ"
            if button:
                button.click()
                break
        except NoSuchElementException:
            scrolled_height = driver.execute_script("return window.scrollY;")
            total_height = driver.execute_script("return document.body.scrollHeight;")
            
            if scrolled_height >= total_height - 1:
                # Đã cuộn đến cuối trang và không tìm thấy nút "Xem thêm"
                print("Đã cuộn đến cuối trang và không tìm thấy nút 'Xem thêm'.")
                break
            continue
    sleep(random.randint(1,4))
    try:
        text = driver.find_element(By.CSS_SELECTOR,".specification-keys").text
        dict_item['feature'] = text
    except:
        dict_item['feature'] = None
    #
    try:
        rating = driver.find_element(By.CSS_SELECTOR,".score .score-average").text
        dict_item['rating'] = rating
    except:
        dict_item['rating'] = None
    list_item.append(dict_item)
```

```python
df_2 = pd.DataFrame(list_item)
df_2
```

```python
df_2.to_csv('rating_final.csv')
```
